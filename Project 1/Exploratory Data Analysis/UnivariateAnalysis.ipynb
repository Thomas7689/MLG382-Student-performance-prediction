```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../Student_performance_data.csv')
df.head()
```

```python
#Histogram of AGe
sns.histplot(data=df, x="Age")
plt.title("Distribution of Student Age")
plt.xlabel("Age(years)")
plt.ylabel("Number of Students")
plt.show()
```

```python
#Histogram of studytime
sns.histplot(data=df, x="StudyTimeWeekly")
plt.title("Weekly Study Time Distribution")
plt.xlabel("StudyTime(hours per week)")
plt.ylabel("Number of Students")
plt.show()
```

```python
#Absences
sns.histplot(data=df, x="Absences")
plt.title("Distribution of School Absences")
plt.xlabel("Number of Absences")
plt.ylabel("Number of Students")
plt.show()
```

```python
#Histogram of GPA
sns.histplot(data=df, x="GPA")
plt.title("GPA Distribution")
plt.xlabel("GPA")
plt.ylabel("Number of Students")
plt.show()
```
