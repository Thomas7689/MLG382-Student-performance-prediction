```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../Student_performance_data.csv')
df.head()
```

```python
#Count plots
for column in ['Gender', 'Ethnicity', 'ParentalEducation', 'Tutoring', 'ParentalSupport', 
               'Extracurricular', 'Sports', 'Music', 'Volunteering', 'GradeClass']:
    plt.figure(figsize=(8, 6))
    sns.countplot(data=df, x=column)
    plt.title(f"Count Plot of {column}")
    plt.xticks(rotation=0)
    plt.show()
```

```python
#Boxplots for numerical values
for column in ['StudyTimeWeekly', 'Absences', 'GPA']:
    plt.figure(figsize=(8,6))
    sns.boxplot(data=df, x='GradeClass', y=column)
    plt.title(f"Boxplot of {column} vs GradeClass")
    plt.show() 
```

```python
#Grouped bar plots
for column in ['Gender', 'Tutoring', 'ParentalSupport']:
    plt.figure(figsize=(8,6))
    sns.countplot(data=df, x=column, hue='GradeClass')
    plt.title(f"Count Plot of {column} vs GradeClass")
    plt.show()
```
